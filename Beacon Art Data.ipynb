```python
import plotly.graph_objects as go
import pandas as pd

data = pd.read_csv('March 27 2022 Datasets - Sheet1 - Copy.csv') # I used 2927 rows not the whole dataset to make it consistent

fig = go.Figure()

for i in range(0, data.shape[1], 2):
    fig.add_trace(go.Scatter3d(
        x=data.iloc[:, i+1],
        y=data.iloc[:, i],
        z=data.index,
        mode='lines',
        name=data.columns[i].split(' ')[0]
    ))

fig.update_layout(
    scene=dict(
        xaxis_title='Longitude',
        yaxis_title='Latitude',
        zaxis_title='Time'
    )
)

fig.show()
```

```python
import plotly.graph_objects as go
import pandas as pd
import plotly.io as pio

data = pd.read_csv('March 27 2022 Datasets - Sheet1 - Copy.csv') # I used 2927 rows not the whole dataset to make it consistent

fig = go.Figure()

for i in range(0, data.shape[1], 2):
    fig.add_trace(go.Scatter3d(
        x=data.iloc[:, i+1],
        y=data.iloc[:, i],
        z=data.index,
        mode='lines',
        name=data.columns[i].split(' ')[0]
    )) 
fig.update_layout(scene=dict(
    xaxis=dict(showgrid=False, zeroline=False, showticklabels=False, ticks='', showbackground=False, title_text=''),
    yaxis=dict(showgrid=False, zeroline=False, showticklabels=False, ticks='', showbackground=False, title_text=''),
    zaxis=dict(showgrid=False, zeroline=False, showticklabels=False, ticks='', showbackground=False, title_text=''),
))

fig.show()
fig.write_image("plot.svg")
fig.write_html("plot.html")
```

```python
import plotly.graph_objects as go
import pandas as pd
import numpy as np
import plotly.io as pio

data = pd.read_csv('March 27 2022 Datasets - Sheet1 - Copy.csv')

fig = go.Figure()

n_entities = int(data.shape[1] / 2)
n_time = len(data)

x_surface, y_surface, z_surface = [], [], []

for t in range(n_time):
    for i in range(n_entities):
        x = data.iloc[t, 2*i+1]
        y = data.iloc[t, 2*i]
        z = t
        
        x_surface.append(x)
        y_surface.append(y)
        z_surface.append(z)

fig.add_trace(go.Surface(x=np.array(x_surface).reshape(n_time, n_entities),
                         y=np.array(y_surface).reshape(n_time, n_entities),
                         z=np.array(z_surface).reshape(n_time, n_entities),
                         colorscale='Blues', # Viridis
                         showscale=False)) 

fig.update_layout(showlegend=False) 
fig.update_layout(scene=dict(
    xaxis=dict(showgrid=False, zeroline=False, showticklabels=False, ticks='', showbackground=False, title_text=''),
    yaxis=dict(showgrid=False, zeroline=False, showticklabels=False, ticks='', showbackground=False, title_text=''),
    zaxis=dict(showgrid=False, zeroline=False, showticklabels=False, ticks='', showbackground=False, title_text=''),
))

fig.show()
fig.write_image("plot2.svg")
fig.write_html("plot2.html")
```
